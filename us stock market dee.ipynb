```python
#creating a copy of a dataframe into  a temporary variable 
data=df[["INDEX", "Date", "Natural_Gas_Price", "Crude_oil_Price", "Tesla_Price", "Apple_Price"]].copy()
```

```python
data.head
```

```python
#extarting feature of year from  date
data['year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data['year'].unique()
```

```python
sns.set_palette('colorblind')
sns.countplot(x ='year', order=[2024, 2023, 2022, 2021, 2020, 2019], hue ='Natural_Gas_Price', data = data)
```

```python
df_grp=data.groupby(['year'])['Natural_Gas_Price'].max().reset_index()
```

```python
df_grp["Crude_oil_Price"]=data.groupby(['year'])["Crude_oil_Price"].max().reset_index()["Crude_oil_Price"]
```

```python
df_grp['Tesla_Price']=data.groupby(['year'])['Tesla_Price'].max().reset_index()['Tesla_Price']
```

```python
df_grp['Apple_Price']=data.groupby(['year'])['Apple_Price'].max().reset_index()['Apple_Price']
```

```python
df_grp
```

```python
sns.barplot(x='year', y='Natural_Gas_Price', hue='Crude_oil_Price', data=df_grp)
```

```python
sns.lineplot(data = df_grp, y='Natural_Gas_Price', x='year', palette='deep')
```

```python
sns.lineplot(data = df_grp, y='Crude_oil_Price', x='year', palette='deep')
```

```python
sns.lineplot(data = df_grp, y='Tesla_Price', x='year', palette='deep')
```

```python
sns.lineplot(data = df_grp, y='Apple_Price', x='year', palette='deep')
```
